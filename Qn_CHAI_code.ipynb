9. Determine if a year is a leap yr (Leap yr are divisible by 4, but not by 100 unless also divisible by 400)

```python
year = 2002

if year % 2 == 0  and year % 100 != 0 or year % 400 == 0:
  print(year,"Leap yr")
else :
  print(year,"Not a leap yr")
```

1. Couting positive numbers
Given a list of numbers count how many num are positive

```python
numbers = [1,4,-3,7,-2,-6,-1,-10]
count = 0

for i in numbers:
  if i < 0:
    count += 1

print("Negative number count:",count)
```

2. sum of even number
calculate the sum of even numbers up to a given number n

```python
n = 20
even_nums_sum = 0
even_number_count = 0
for i in range(1,n+1):
  if i % 2 == 0:
    even_nums_sum += i
    even_number_count += 1
print("Even num sum:",even_nums_sum, "\nEven num count:",even_number_count)
```

6. Factorial of a number using while loop

```python
num = 5
factorial = 1

while num > 0:
  factorial = factorial* num
  num = num - 1
print(factorial)
```

Check if the num is prime

```python
num = int(input("Enter a number:"))
while num > 1 :
  for i in range (2,num):
    if num % i == 0:
      print("Not a prime number")
      break
    else:
      print("Prime Number")
      break
  break
```
