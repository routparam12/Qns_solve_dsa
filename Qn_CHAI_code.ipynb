Fruit ripeness checker
Determine if a fruit is ripe, overripe,or unique based on its color (Banana:Green - unripen, yellow - ripen, brown - overripen)



```python
fruit = "Banana"
color = "Yellow"
if fruit == "Banana":
  if color == "Green":
    print("Unripen")
  elif color == "Yellow":
    print("Ripe")
  else :
    print("Over ripe")
```

8. If password is weak or medium or strong. Criteria < 6 char(weak), (6-10 char)(Medium), > 10 Char (Strong)

```python
Password = 'PPPPPPPP'

if len(Password) < 6:
  strength = "Weak"
elif 6 <= len(Password) < 10:
  strength = "Medium"
elif len(Password) >= 10:
  strength =  "Strong"

print(strength)
```

9. Determine if a year is a leap yr (Leap yr are divisible by 4, but not by 100 unless also divisible by 400)

```python
year = 2002

if year % 2 == 0  and year % 100 != 0 or year % 400 == 0:
  print(year,"Leap yr")
else :
  print(year,"Not a leap yr")
```
