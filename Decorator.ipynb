# Learning about decorator
Qn: Write a decorator that measures the time a function takes to execute

```python
import time

def timer(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs) # func is example_function, its return value is stored here
        end = time.time()
        print(f"{func.__name__} ran in {end-start} time")
        print(result)
        return result # The wrapper returns the result of the original func
    return wrapper

@timer
def example_function(n):
  time.sleep(n)
  return f"Successfully slept for {n} seconds!" # Uncommented to return a specific value
  # return result # This line would refer to a global 'result' variable


# Call the decorated function and capture its return value
returned_message = example_function(2)
print(f"The decorated function returned: {returned_message}")
```

```python
def my_function(**kwargs):
    print("Received keyword arguments:")
    for key, value in kwargs.items():
        print(f"  {key}: {value}")
    return

# Calling the function with different keyword arguments
my_function(name="Alice", age=30, city="New York")
print("\n---")
my_function(product="Laptop", price=1200)
```

```python
# Define a simple function that takes two arguments
def add_numbers(a, b):
    sum_result = a + b
    return sum_result

# Call the function with specific arguments
result = add_numbers(5, 4)
print(f"The sum is: {result}")

# Call the function again with different arguments
result_2 = add_numbers(10, 20)
print(f"The second sum is: {result_2}")
```

```python
def my_function_with_return(**kwargs):
    print("Received keyword arguments:")
    for key, value in kwargs.items():
        print(f"  {key}: {value}")
    # Return the dictionary of keyword arguments
    return kwargs

# Call the function and store the returned dictionary
returned_args_1 = my_function_with_return(name="Alice", age=30, city="New York")
print(f"Returned value: {returned_args_1}")

print("\n---")

returned_args_2 = my_function_with_return(product="Laptop", price=1200, brand="Acme")
print(f"Returned value: {returned_args_2}")
```

Qn 2: Create a decorator to print the function name and the values of its arguments every time the function is called

```python
def debug_toll(func):
  def wrapper(*args, **kwargs):
    args_value = ', '.join(str(arg) for arg in args )
    kwargs_value = ', '.join(f"{k}={v}" for k,v in kwargs.items())
    print(f"Calling {func.__name__} with arguments: {args_value}, {kwargs_value}")
    return func(*args, **kwargs)
  return wrapper

@debug_toll
def greet(name, greeting= "Namaste"):
  print(f"{greeting}, {name}")

greet("Param",greeting="Good morning")
greet('Param')
```

```python
def my_function_with_return(first_arg, **kwargs):
    print(f"First positional argument: {first_arg}")
    print("Received keyword arguments:")
    for key, value in kwargs.items():
        print(f"  {key}: {value}")
    return kwargs

# Call the function with an additional positional argument and keyword argument
returned_args_3 = my_function_with_return("John", name="John", age=40, city="London", occupation="Engineer")
print(f"Returned value: {returned_args_3}")

print("\n---")

returned_args_4 = my_function_with_return("Data", product="Keyboard", price=75, color="Black", supplier="Logi")
print(f"Returned value: {returned_args_4}")
```

```python
def show_me_the_data(*args, **kwargs):
    print(f"Arguments in *args (Tuple): {args}")
    print(f"Arguments in **kwargs (Dictionary): {kwargs}")

# Calling the function with a mix of both
print("--- Execution Output ---")
show_me_the_data(1, "Param", 3, name="Python", version=3.12, status="Learning")
```

```python
def example_collector(arg1, arg2, *additional_args, **additional_kwargs):
    print(f"Named Positional Argument 1 (arg1): {arg1}")
    print(f"Named Positional Argument 2 (arg2): {arg2}")
    print(f"*additional_args (Tuple): {additional_args}")
    print(f"**additional_kwargs (Dictionary): {additional_kwargs}")

print("--- Call 1: With named args, some positional, and some keyword ---")
example_collector("value1", "value2", 10, 20, item="apple", count=5)

print("\n--- Call 2: With named args, more positional, and more keyword ---")
example_collector("first", "second", True, 3.14, 'hello', color="blue", size="large", active=False)
```

```python
# Define a decorator
def my_decorator(func):
    def wrapper():
        print("Before the function runs")
        func()
        print("After the function runs")
    return wrapper

# Apply decorator
@my_decorator
def say_hello():
    print("Hello, World!")
@my_decorator
def sayyy_hello():
    print("Hello, Worllld!")

say_hello()
sayyy_hello()
```
