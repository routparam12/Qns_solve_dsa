# Learning about decorator
Qn: Write a decorator that measures the time a function takes to execute

```python

import time

def timer(func):
    def wrapper(*args, **kwargs):
        start = time.time()
        result = func(*args, **kwargs)
        end = time.time()
        print(f"{func.__name__} ran in {end-start} time")
        return result
    return wrapper

@timer
def example_function(n):
  time.sleep(n)
example_function(2)

```

```python
def my_function(**kwargs):
    print("Received keyword arguments:")
    for key, value in kwargs.items():
        print(f"  {key}: {value}")
    return

# Calling the function with different keyword arguments
my_function(name="Alice", age=30, city="New York")
print("\n---")
my_function(product="Laptop", price=1200)
```

```python
# Define a simple function that takes two arguments
def add_numbers(a, b):
    sum_result = a + b
    return sum_result

# Call the function with specific arguments
result = add_numbers(5, 3)
print(f"The sum is: {result}")

# Call the function again with different arguments
result_2 = add_numbers(10, 20)
print(f"The second sum is: {result_2}")
```

```python
def my_function_with_return(**kwargs):
    print("Received keyword arguments:")
    for key, value in kwargs.items():
        print(f"  {key}: {value}")
    # Return the dictionary of keyword arguments
    return kwargs

# Call the function and store the returned dictionary
returned_args_1 = my_function_with_return(name="Alice", age=30, city="New York")
print(f"Returned value: {returned_args_1}")

print("\n---")

returned_args_2 = my_function_with_return(product="Laptop", price=1200, brand="Acme")
print(f"Returned value: {returned_args_2}")
```

```python
# Define a decorator
def my_decorator(func):
    def wrapper():
        print("Before the function runs")
        func()
        print("After the function runs")
    return wrapper

# Apply decorator
@my_decorator
def say_hello():
    print("Hello, World!")
@my_decorator
def sayyy_hello():
    print("Hello, Worllld!")

say_hello()
sayyy_hello()
```
