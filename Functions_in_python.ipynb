Create a function thaat returns both the area and circumstances of a circle given its radius.

```python
def radius(r):
  radius = 3.14 * (r**2)
  circumference = 2 * 3.14 * r
  print("Radius:", radius,"And Circumference:",circumference)

radius(5)
```

```python
import math
def radius(r):
  radius = math.pi * (r**2)
  circumference = 2 * math.pi * r
  print("Radius:", round(radius,2),"And Circumference:",round(circumference,2))

radius(5)
```
